```python
import argparse
from Bio import SeqIO

# Lista de especies (puedes guardarla en otro archivo si prefieres)
species_list = list(df['species'].unique())


# Función de búsqueda
def buscar_especie_en_fna(fna_path):
    record = next(SeqIO.parse(fna_path, "fasta"))
    header = record.description.lower()
    
    matches = []
    for species in species_list:
        especie = species.replace("s__", "").lower()
        if especie in header:
            matches.append(species)

    df_matches = df[df['species'] == matches[0]]
    
    if matches:
        print(f"✅ Coincidencias encontradas en el header:\n{matches}")
    else:
        print("❌ No se encontraron especies reconocidas en el header.")
        
    return header, matches, df_matches
```

```python
# Cambia esto por la ruta a tu archivo .fna
ruta_fna = "inputs/GCA_000007385.1_ASM738v1_genomic.fna"

# Ejecuta búsqueda
header, coincidencias, df_char = buscar_especie_en_fna(ruta_fna)
```
